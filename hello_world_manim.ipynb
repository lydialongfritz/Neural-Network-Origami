```python
from manim import *
```

```python
%%manim -ql -v WARNING SingleNeuron

class SingleNeuron(ThreeDScene):
    def construct(self):
        axes_3d = ThreeDAxes(
            x_range=(-6,6,1),
            x_length=12,
            y_range=(-5,5,1),
            y_length=10,
            z_range=(-3,3,1),
            z_length=6
        )
        self.add(axes_3d)
        self.set_camera_orientation(phi=60*DEGREES, theta=-45*DEGREES)

        u_weight_tracker = ValueTracker(4)
        v_weight_tracker = ValueTracker(1)

        def compute_location(u, v):
            u_weight = u_weight_tracker.get_value()
            v_weight = v_weight_tracker.get_value()
            z = max((u_weight * u) + (v_weight * v) - 3, 0)
            return np.array([u,v,z])

        neuron_plane = Surface(
            compute_location,
            u_range=[-6,6],
            v_range=[-5,5]
        ).set_opacity(0.8)

        def update_surface(mob):
            new_surface = Surface(
                compute_location,
                u_range=[-6,6],
                v_range=[-5,5]
            ).set_opacity(0.8)
            mob.become(new_surface)

        self.add(
            axes_3d,neuron_plane
        )

        neuron_plane.add_updater(update_surface)

        self.move_camera(phi=75*DEGREES, theta=-70*DEGREES,run_time=2)

        # change the weights
        self.play(u_weight_tracker.animate.set_value(2), run_time=3)

        # add the points
        point = Dot3D(point=compute_location(1,0), color=BLUE_D, radius=0.12)
        point_shadow = Dot3D(point=[1,0,0], color=0x405b61, radius=0.12)

        self.add(point)
        self.add(point_shadow)

        # move the points along the surface
        self.play(
            point.animate.move_to(compute_location(3,0)),
            point_shadow.animate.move_to([3,0,0]),
            run_time=4)
        self.play(
            point.animate.move_to(compute_location(1,4)),
            point_shadow.animate.move_to([1,4,0]),
            run_time=4)
        self.play(
            point.animate.move_to(compute_location(1,0)),
            point_shadow.animate.move_to([1,0,0]),
            run_time=4)

        self.wait(2)
```
